```python
import pandas as pd
data=pd.read_csv('data/d_train_20180102.csv')
from sklearn.preprocessing import LabelEncoder
import numpy as np
print(np.unique(data['性别'].values))
LE=LabelEncoder()
sex=LE.fit_transform(data['性别'].values)
data['性别']=sex
data.pop('体检日期')
data.pop('id')
nan_1={'乙肝表面抗原':0,'乙肝表面抗体':0,'乙肝e抗原':0,'乙肝e抗体':0,'乙肝核心抗体':0}
data.fillna(value=nan_1,inplace=True)
X_train=data.values[:,0:-1]
y_train=data.values[:,-1]

feat_names=list(data.columns)
feat_names.remove('血糖')
print(feat_names)
data.head()
```

```python
data_test=pd.read_csv('data/d_test_A_20180102.csv')
sex=LE.fit_transform(data_test['性别'].values)
data_test['性别']=sex
data_test.pop('体检日期')
data_test.pop('id')
nan_1={'乙肝表面抗原':0,'乙肝表面抗体':0,'乙肝e抗原':0,'乙肝e抗体':0,'乙肝核心抗体':0}
data_test.fillna(value=nan_1,inplace=True)
X_test=data_test.values
data_test.head()
```

```python
import lightgbm as lgb
from sklearn.model_selection import ParameterGrid
import matplotlib.pyplot as plt
params = {
    'metric': ['mse'],
    'objective': ['regression'],
    'learning_rate':[0.15],
    'feature_fraction': [0.8],
    'max_depth': [4],
    'num_leaves':[80],
    'bagging_fraction': [0.8],
    'bagging_freq':[5],
    'min_data_in_leaf':[1],
    'min_gain_to_split':[0],
    'lambda_l1':[1],
    'lambda_l2':[1],
    'verbose':[0]
}
params=list(ParameterGrid(params))
lgbtrain=lgb.Dataset(X_train,label=y_train,feature_name=feat_names,categorical_feature=[0])
for param in params:
    print(param)
    clf = lgb.train(param, lgbtrain, num_boost_round=59, categorical_feature=[0])
result=clf.predict(X_test)
print(result.shape)
print(result)
result=pd.DataFrame(result)
result.to_csv('result.csv',index=None,columns=None,header=None)
```
